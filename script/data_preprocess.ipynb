```python
from hanziconv import HanziConv
datafiles = ['ming.all', 'qing.all', 'qss_tab.txt', 'qts_tab.txt', 'qsc_tab.txt', 'qtais_tab.txt', 'yuan.all']
terminate_symbols = [r'。',r'！',r'？',r'\?',r'!']
all_sentences = []
for filename in datafiles:
    df  = pd.read_csv(DATA_DIR+filename, sep='\t')
    for idx in range(df.shape[0]):
        body = df.loc[idx, 'body']
        body = HanziConv.toSimplified(body)
        body = body.encode('utf-8')
        all_sentences.append(body)
        for s in terminate_symbols:
            body.strip()
            body = re.sub(s, '{}\\\\'.format(s), body)
        body.strip()
        sentences = body.split("\\")
```

```python
# build vocab
import gensim
import jieba
all_seg_sentences = []
for idx in range(len(all_sentences)):
    s = all_sentences[idx]
    if type(s) != unicode:
        s = s.decode('utf-8')
    seg = jieba.tokenize(s)
    all_seg_sentences.append([w[0] for w in seg])
```

```python
all_seg_sentences1 = [[w for w in s.decode('utf-8')] for s in all_sentences]
```

```python
from gensim.corpora import Dictionary as corpora_Dictionary
vocab_output_file = 'work/vocab.txt'
output_file = 'work/doc_idx.txt'
dct = corpora_Dictionary(all_seg_sentences1)
with open(vocab_output_file, 'wb') as f:
    for idx in range(len(dct)):
        f.write(("%d\\%s\\%d\n"%(idx, dct[idx], dct.dfs[idx])).encode("utf-8"))

all_seg_idx = []

for idx in range(len(all_seg_sentences1)):
    idx_sentences = dct.doc2idx(all_seg_sentences1[idx])
    all_seg_idx.append(idx_sentences)

with open(output_file, 'wb') as f:
    for idx in range(len(all_seg_idx)):
        s = all_seg_idx[idx]
        f.write(("{}\n".format(" ".join([str(x) for x in s]))).encode("utf-8"))
```

```python
# keep frequence more than 1 words
from gensim.corpora import Dictionary as corpora_Dictionary
vocab_output_file = 'work/vocab_feq_gt1.txt'
output_file = 'work/doc_idx_feq_gt1.txt'
dct = corpora_Dictionary(all_seg_sentences1)
dct.filter_extremes(no_below=2, no_above=1, keep_n=10000000)
with open(vocab_output_file, 'wb') as f:
    # rear word
    f.write(("0\\S\\1\n").encode("utf-8"))
    f.write(("1\\B\\1\n").encode("utf-8"))
    for idx in range(len(dct)):
        f.write(("%d\\%s\\%d\n"%(idx+2, dct[idx], dct.dfs[idx])).encode("utf-8"))

all_seg_idx = []

for idx in range(len(all_seg_sentences1)):
    idx_sentences = dct.doc2idx(all_seg_sentences1[idx])
    all_seg_idx.append(idx_sentences)

with open(output_file, 'wb') as f:
    for idx in range(len(all_seg_idx)):
        s = all_seg_idx[idx]
        f.write(("{}\n".format(" ".join([str(x+1) for x in s]))).encode("utf-8"))
```
